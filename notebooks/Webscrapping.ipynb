```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import time
import re
from tqdm.notebook import tqdm
```

```python
dict_states = {'AL': 7, 'AK': 1, 'AZ': 9, 'AR': 4, 'CA': 53, 'CO': 7, 'CT': 5, 'DE': 1, 
               'FL': 27, 'GA': 14, 'HI': 2, 'ID': 2, 'IL': 18, 'IN': 9, 'IA': 4, 'KS': 4,
              'KY': 6, 'LA': 6, 'ME': 2, 'MD': 8, 'MA': 9, 'MI': 13, 'MN': 8, 'MS': 4, 'MO': 8,
              'MT': 1, 'NE': 3, 'NV': 4, 'NH': 2, 'NJ': 12, 'NM': 3, 'NY': 27, 'NC': 13, 
              'ND': 1, 'OH': 16, 'OK': 5, 'OR': 5, 'PA': 18, 'RI': 2, 'SC': 7, 'SD': 1, 
              'TN': 9, 'TX': 34, 'UT': 4, 'VT': 1, 'VA': 11, 'WA': 10, 'WI': 8, 'WV': 3, 'WY': 1}                
#use dictionary
def generate_URL(state):
    """Function turns the dictionary of states: and number of congressional districts """
    URL_list = []
    for i in range(1, dict_states[state] + 1):
        URL = "https://www.opensecrets.org/races/summary?cycle=2020&id=" + state + str(i).zfill(2) + "&spec=N"
        URL_list.append(URL)
    return URL_list

# generates a list of the URLs associated with the data for congressional districts
# URL_list = []
results_df = []
for state in tqdm(dict_states):
    URL_list = generate_URL(state)

#results_df = []

#loop generates a list of dataframes
    for URL in URL_list:
        time.sleep(0.1)
        page = requests.get(URL)
        df = pd.read_html(page.text)
        df = df[0]
        df['state'] = re.findall(r"\b[A-Z]{2}\d{2}", URL)[0]
        results_df.append(df)

results_df
```

```python
final_df = pd.concat(results_df)
final_df
```
