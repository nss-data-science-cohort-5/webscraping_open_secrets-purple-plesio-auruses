```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_districts_us = (pd.read_csv('all_districts_us').
                    drop(columns = 'Unnamed: 0'))
```

```python
all_districts_us['winner'] = all_districts_us['winner'].astype(bool)
all_districts_us['incumbent'] = all_districts_us['incumbent'].astype(bool)
```

# How often does the candidate who raised more money win a race?

```python
all_districts_us['Raised'] = all_districts_us['Raised'].str.replace('$','').str.replace(',','').astype(int)
all_districts_us['Spent'] = all_districts_us['Spent'].str.replace('$','').str.replace(',','').astype(int)
```

```python
fig = plt.subplots(figsize = (15,10))
plt.bar(x = all_districts_us['state'].drop_duplicates(),
        height = all_districts_us.groupby('state')['Raised'].mean())
plt.ylabel('Average money raised per state');
```
